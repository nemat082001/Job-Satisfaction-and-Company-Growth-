```python
import pandas as pd
from sklearn.metrics import jaccard_score

# Sample data: df['text'] and top 10 words for each of the 6 topics
# Replace this with your actual data
topics_words = [
            ['work', 'good', 'life', 'balance', 'hike', 'salary', 'less', 'growth', 'opportunity', 'low'],
            ['get', 'manager', 'employee', 'year', 'management', 'work', 'bad', 'even', 'time', 'people'],
            ['good', 'work', 'salary', 'hike', 'less', 'pay', 'hour', 'shift', 'employee', 'flexible'],
            ['project', 'good', 'learn', 'technology', 'get', 'new', 'opportunity', 'work', 'lot', 'freshers'],
            ['employee', 'process', 'management', 'growth', 'opportunity', 'good', 'level', 'policy', 'carrer', 'lack'],
            ['good', 'work', 'enviornment', 'place', 'con', 'great', 'nice', 'culture', 'team', 'friendly']
]

# Topic-Word Distribution: Create a binary matrix for top words in each topic
topic_word_matrix = pd.DataFrame(0, index=range(len(topics_words)), columns=list(topics_words))

topic_word_matrix = pd.DataFrame(0, index=range(len(topics_words)), columns=topics_words)
for i, words in enumerate(topics_words):
    for word in words:
        topic_word_matrix.loc[i, word] = 1
        
# Document-Topic Distribution: Calculate presence of top words for each topic in documents
# Document-Topic Distribution: Calculate presence of top words for each topic in documents
document_topic_matrix = pd.DataFrame(0, index=range(len(df)), columns=range(len(topics_words)))
for i, text in enumerate(df['text']):
    words_in_text = set(text.split())
    for j, topic_words in enumerate(topics_words):
        presence = any(word in words_in_text for word in topic_words)
        document_topic_matrix.loc[i, j] = 1 if presence else 0

        
document_topic_matrix = document_topic_matrix.fillna(0)
topic_word_matrix = topic_word_matrix.fillna(0)
# Calculate Jaccard similarity score between documents and topics
similarity_scores = []
for i in range(len(df)):
    document_vector = document_topic_matrix.loc[i].values
    topic_vectors = topic_word_matrix.values
    similarity = jaccard_score(document_vector, topic_vectors, average=None)
    similarity_scores.append(similarity)

# Create columns for each topic and store the similarity scores
for i in range(len(topics_words)):
    df[f'topic_{i + 1}_score'] = similarity_scores[i]
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Sample topics and words
topics_words = [
    ['work', 'good', 'life', 'balance', 'hike', 'salary', 'less', 'growth', 'opportunity', 'low'],
    ['get', 'manager', 'employee', 'year', 'management', 'work', 'bad', 'even', 'time', 'people'],
    ['good', 'work', 'salary', 'hike', 'less', 'pay', 'hour', 'shift', 'employee', 'flexible'],
    ['project', 'good', 'learn', 'technology', 'get', 'new', 'opportunity', 'work', 'lot', 'freshers'],
    ['employee', 'process', 'management', 'growth', 'opportunity', 'good', 'level', 'policy', 'carrer', 'lack'],
    ['good', 'work', 'environment', 'place', 'con', 'great', 'nice', 'culture', 'team', 'friendly']
]

# Extract unique words from topics
unique_words = list(set(word for sublist in topics_words for word in sublist))

# Create a CountVectorizer with the predefined vocabulary (topics_words)
vectorizer = CountVectorizer(vocabulary=unique_words)

# Transform the text data into a document-term matrix
dtm = vectorizer.fit_transform(df['text'])

# Calculate topic probabilities
topic_probabilities = dtm.dot(vectorizer.transform(sum(topics_words, [])).T)

# Normalize topic probabilities to sum up to 1 for each document
normalized_topic_probabilities = topic_probabilities / topic_probabilities.sum(axis=1)

# Create columns for each topic and store the probabilities
num_topics_to_keep = 6
for i in range(num_topics_to_keep):
    df[f'topic_{i + 1}'] = normalized_topic_probabilities[:, i]
```
